```python
with open('expanded_intents.json') as f:
  raw_data = json.load(f)
```

```python
raw_data['intents']
```

```python
print("number of intents =", len(raw_data['intents']))
for intent in raw_data['intents']:
    print("Intent", intent['intent'] , len(intent['text']))
```

```python
response_data = {
    "intents": []
}
```

```python
for intent in raw_data['intents']:
  if len(intent['text']) >= 40:
    response_data['intents'].append(
        {
            'tag': intent['intent'],
            'responses': intent['responses']
        }
    )
  else:
    print(intent['intent'])
```

```python
len(response_data['intents'])
```

```python
train_data = {
    'text': [],
    'label': []
}
```

```python
for intent in raw_data['intents']:
  if len(intent['text']) >= 40:
    for text in intent['text']:
      train_data['text'].append(text)
      train_data['label'].append(intent['intent'])
```

```python
train_df = pd.DataFrame(train_data)
train_df
```

```python
with open('intents_responses.json', 'w') as f:
  json.dump(response_data, f)
```

```python
train_df.to_csv('train_data.csv', index=False)
```
